```python
import pandas as pd

# Column names
cols = [
    "state", "ssa", "county_name", "parta_enroll",
    "parta_reimb", "parta_percap", "parta_reimb_unadj",
    "parta_percap_unadj", "parta_ime", "parta_dsh",
    "parta_gme", "parta_demo", "partb_enroll",
    "partb_reimb", "partb_percap", "partb_demo",
    "mean_risk"
]

# Read CSV
ffs_data = pd.read_csv(
    "data/input/ffs-costs/Extracted Data/Aged Only/AGED08.csv",
    skiprows=4,
    names=cols,
    na_values="*"
)

# Explicit type conversion (equivalent to col_types)
ffs_data["state"] = ffs_data["state"].astype("string")
ffs_data["county_name"] = ffs_data["county_name"].astype("string")

numeric_cols = [
    "ssa", "parta_enroll", "parta_reimb", "parta_percap",
    "parta_reimb_unadj", "parta_percap_unadj", "parta_ime",
    "parta_dsh", "parta_gme", "parta_demo",
    "partb_enroll", "partb_reimb", "partb_percap",
    "partb_demo", "mean_risk"
]

ffs_data[numeric_cols] = ffs_data[numeric_cols].apply(
    pd.to_numeric, errors="coerce"
)

# Select columns + mutate year
final_ffs_costs = (
    ffs_data[
        ["ssa", "state", "county_name",
         "parta_enroll", "parta_reimb",
         "partb_enroll", "partb_reimb", "mean_risk"]
    ]
    .assign(year=2008)
)
```
